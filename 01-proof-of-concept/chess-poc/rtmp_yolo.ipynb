```python
import cv2
from ultralytics import YOLO
from huggingface_hub import hf_hub_download

weights_path = hf_hub_download(
    repo_id="dopaul/chess-piece-detector-merged",
    filename="best.pt",
    local_dir="./models",
)

# Load your chess-piece model
model = YOLO(weights_path)            # path to dopaul/chess-board-segmentation weights

# RTMP URL (local)
stream = "rtmp://localhost/live/chess"
cap = cv2.VideoCapture(stream)

frame_i = 0
while cap.isOpened():
    ok, frame = cap.read()
    if not ok:
        continue      # wait until stream is live

    # 4️⃣  Run inference every 10th frame
    if frame_i % 10 == 0:
        results = model(frame)

        # 5️⃣  Print detections (optional)
        for box, cls, conf in zip(results[0].boxes.xyxy,
                                  results[0].boxes.cls,
                                  results[0].boxes.conf):
            piece = model.names[int(cls)]
            print(f"{piece:9s}  conf={conf:.2f}  box={box.cpu().numpy()}")

        # 6️⃣  Show annotated frame
        cv2.imshow("Detections", results[0].plot())   # Ultralytics handles BGR→RGB
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    frame_i += 1

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

model = YOLO("best.pt")  # update this path if needed
cap = cv2.VideoCapture("rtmp://localhost/live/chess")

frame_i = 0
while cap.isOpened():
    ok, frame = cap.read()
    if not ok:
        print("Stream lost or not yet available...")
        continue

    if frame_i % 10 == 0:
        results = model(frame)
        cv2.imshow("Detections", results[0].plot())
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    frame_i += 1

cap.release()
cv2.destroyAllWindows()
```
