```python
# ndi_capture.py
import cv2, ndi  # pip install ndi-python
from ultralytics import YOLO

ndi.initialize()
recv = ndi.Recv()
sources = ndi.find_sources()
if not sources:
    raise RuntimeError("No NDI sources found. Make sure the phone app is on.")
recv.connect(sources[0])

model = YOLO("best.pt")             # path to the chess weights
frame_i = 0

while True:
    video_frame = recv.capture_video(timeout_ms=5000)
    if video_frame is None:
        continue  # lost frame, skip
    frame = video_frame.data  # numpy BGR
    if frame_i % 10 == 0:
        results = model(frame)
        cv2.imshow("chess", results[0].plot())
        if cv2.waitKey(1) == ord("q"):
            break
    frame_i += 1
```
